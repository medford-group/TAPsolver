# Running the forward TAP problem and visualizing the results

```python
from tapsolver import *
```

Now that we've defined the TAPobject, we can start playing around. First, let's load the TAPobject we generated in our first example.

```python
tap_1 = read_TAPobject('./example_TAPobject.json')
```

Next, we can simply run a single pulse experiment. This is done by calling the forward_problem function.

```python
# The first input is the time of the simulation (seconds), the second input is the number of pulses, and the third input is the TAPobject

forward_problem(1,1,tap_1)
```

We can visualize this simulation with the flux_graph command.

```python
flux_graph(tap_1)
```

To control the amount of noise in our simulation, we can simply adjust it in the TAPobject for each gas. So let's increase the noise for argon and decrease it for propane

```python
tap_1.reactor_species.gasses['C3H8'].noise = 0.01
tap_1.reactor_species.inert_gasses['Ar'].noise = 0.1
```

```python
forward_problem(1,1,tap_1)
flux_graph(tap_1)
```
